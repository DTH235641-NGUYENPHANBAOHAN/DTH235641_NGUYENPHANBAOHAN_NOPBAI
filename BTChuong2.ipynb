```python
#C√¢u 1
def main():
    try:
        # Nh·∫≠p s·ªë ti·ªÅn g·ªëc t·ª´ ng∆∞·ªùi d√πng
        X = float(input("Nh·∫≠p s·ªë ti·ªÅn g·ªëc (X): "))

        # Ki·ªÉm tra s·ªë ti·ªÅn h·ª£p l·ªá
        if X <= 0:
            print("S·ªë ti·ªÅn ph·∫£i l·ªõn h∆°n 0!")
            return

        # Thi·∫øt l·∫≠p c√°c tham s·ªë
        lai_suat_thang = 0.006  # 0.6% = 0.006
        so_thang = 18
        ky_han = 6  # C·ª© 6 th√°ng c·ªông l√£i v√†o g·ªëc m·ªôt l·∫ßn

        # T√≠nh s·ªë k·ª≥ h·∫°n trong 18 th√°ng
        so_ky_han = so_thang // ky_han

        # T√≠nh t·ªïng ti·ªÅn sau m·ªói k·ª≥ h·∫°n
        tong_tien = X

        for ky in range(so_ky_han):
            # T√≠nh l√£i su·∫•t cho 6 th√°ng
            lai_suat_6_thang = (1 + lai_suat_thang) ** ky_han - 1

            # T√≠nh ti·ªÅn l√£i cho k·ª≥ n√†y
            tien_lai = tong_tien * lai_suat_6_thang

            # C·ªông l√£i v√†o g·ªëc
            tong_tien += tien_lai

            print(f"Sau {ky_han * (ky + 1)} th√°ng: {tong_tien:,.0f} VND")

        print("\n" + "="*50)
        print(f"S·ªë ti·ªÅn g·ªëc ban ƒë·∫ßu: {X:,.0f} VND")
        print(f"L√£i su·∫•t th√°ng: {lai_suat_thang * 100}%")
        print(f"Sau {so_thang} th√°ng, t·ªïng s·ªë ti·ªÅn l√†: {tong_tien:,.0f} VND")

    except ValueError:
        print("Vui l√≤ng nh·∫≠p s·ªë ti·ªÅn h·ª£p l·ªá!")

if __name__ == "__main__":
    main()
```

```python
#C√¢u 2
def main():
    try:
        # Nh·∫≠p hai s·ªë t·ª´ ng∆∞·ªùi d√πng
        a = float(input("Nh·∫≠p s·ªë th·ª© nh·∫•t (a): "))
        b = float(input("Nh·∫≠p s·ªë th·ª© hai (b): "))

        # Nh·∫≠p k√Ω t·ª± ph√©p to√°n
        ch = input("Nh·∫≠p k√Ω t·ª± ph√©p to√°n (+, -, *, /): ")

        # Ki·ªÉm tra v√† th·ª±c hi·ªán ph√©p to√°n t∆∞∆°ng ·ª©ng
        if ch == '+':
            ket_qua = a + b
            print(f"{a} + {b} = {ket_qua}")
        elif ch == '-':
            ket_qua = a - b
            print(f"{a} - {b} = {ket_qua}")
        elif ch == '*':
            ket_qua = a * b
            print(f"{a} * {b} = {ket_qua}")
        elif ch == '/':
            if b == 0:
                print("L·ªói: Kh√¥ng th·ªÉ chia cho 0!")
            else:
                ket_qua = a / b
                print(f"{a} / {b} = {ket_qua}")
        else:
            print(f"K√Ω t·ª± '{ch}' kh√¥ng ph·∫£i l√† m·ªôt to√°n t·ª≠ h·ª£p l·ªá")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")

if __name__ == "__main__":
    main()
```

```python
#C√¢u 3
import math

def tinh_dien_tich_hinh_tron():
    try:
        # Nh·∫≠p b√°n k√≠nh
        ban_kinh = float(input("Nh·∫≠p b√°n k√≠nh h√¨nh tr√≤n: "))

        # Ki·ªÉm tra b√°n k√≠nh h·ª£p l·ªá
        if ban_kinh <= 0:
            print("B√°n k√≠nh ph·∫£i l√† s·ªë d∆∞∆°ng!")
            return

        # T√≠nh di·ªán t√≠ch h√¨nh tr√≤n
        dien_tich = math.pi * (ban_kinh ** 2)

        # In k·∫øt qu·∫£
        print("\n" + "="*40)
        print(f"B√°n k√≠nh: {ban_kinh}")
        print(f"Di·ªán t√≠ch h√¨nh tr√≤n: {dien_tich:.2f}")
        print(f"Di·ªán t√≠ch (l√†m tr√≤n): {round(dien_tich, 2)}")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")

if __name__ == "__main__":
   tinh_dien_tich_hinh_tron()
```

```python
#C√¢u 4
def kiem_tram_nam_nhuan(nam):
    """Ki·ªÉm tra nƒÉm nhu·∫≠n"""
    if (nam % 400 == 0) or (nam % 100 != 0 and nam % 4 == 0):
        return True
    return False

def so_ngay_trong_thang():
    """Ch∆∞∆°ng tr√¨nh ch√≠nh t√≠nh s·ªë ng√†y trong th√°ng"""
    try:
        # Nh·∫≠p th√°ng t·ª´ ng∆∞·ªùi d√πng
        thang = int(input("Nh·∫≠p th√°ng (1-12): "))

        # Ki·ªÉm tra th√°ng h·ª£p l·ªá
        if thang < 1 or thang > 12:
            print("Th√°ng ph·∫£i t·ª´ 1 ƒë·∫øn 12!")
            return

        # C√°c th√°ng c√≥ 31 ng√†y
        thang_31_ngay = [1, 3, 5, 7, 8, 10, 12]

        # C√°c th√°ng c√≥ 30 ng√†y
        thang_30_ngay = [4, 6, 9, 11]

        if thang in thang_31_ngay:
            print(f"Th√°ng {thang} c√≥ 31 ng√†y")
        elif thang in thang_30_ngay:
            print(f"Th√°ng {thang} c√≥ 30 ng√†y")
        else:  # Th√°ng 2
            # Y√™u c·∫ßu nh·∫≠p nƒÉm ƒë·ªÉ ki·ªÉm tra nƒÉm nhu·∫≠n
            nam = int(input("Th√°ng 2 - Vui l√≤ng nh·∫≠p nƒÉm: "))

            if kiem_tram_nam_nhuan(nam):
                print(f"Th√°ng 2 nƒÉm {nam} c√≥ 29 ng√†y (nƒÉm nhu·∫≠n)")
            else:
                print(f"Th√°ng 2 nƒÉm {nam} c√≥ 28 ng√†y (nƒÉm kh√¥ng nhu·∫≠n)")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")

if __name__ == "__main__":
    so_ngay_trong_thang()
```

```python
#C√¢u 5
def doc_so_co_2_chu_so():
    """Ch∆∞∆°ng tr√¨nh ƒë·ªçc s·ªë c√≥ t·ªëi ƒëa 2 ch·ªØ s·ªë"""
    try:
        # Nh·∫≠p s·ªë t·ª´ ng∆∞·ªùi d√πng
        n = int(input("Nh·∫≠p m·ªôt s·ªë (0-99): "))

        # Ki·ªÉm tra s·ªë h·ª£p l·ªá
        if n < 0 or n > 99:
            print("S·ªë ph·∫£i t·ª´ 0 ƒë·∫øn 99!")
            return

        # Danh s√°ch c√°ch ƒë·ªçc c√°c s·ªë
        don_vi = ["kh√¥ng", "m·ªôt", "hai", "ba", "b·ªën", "nƒÉm", "s√°u", "b·∫£y", "t√°m", "ch√≠n"]
        chuc = ["", "m∆∞·ªùi", "hai m∆∞∆°i", "ba m∆∞∆°i", "b·ªën m∆∞∆°i", "nƒÉm m∆∞∆°i",
                "s√°u m∆∞∆°i", "b·∫£y m∆∞∆°i", "t√°m m∆∞∆°i", "ch√≠n m∆∞∆°i"]

        # X·ª≠ l√Ω c√°c tr∆∞·ªùng h·ª£p ƒë·∫∑c bi·ªát
        if n == 0:
            print("Kh√¥ng")
            return

        # T√°ch ch·ªØ s·ªë h√†ng ch·ª•c v√† h√†ng ƒë∆°n v·ªã
        hang_chuc = n // 10
        hang_don_vi = n % 10

        # ƒê·ªçc s·ªë theo c√°c tr∆∞·ªùng h·ª£p
        if hang_chuc == 0:
            # S·ªë c√≥ 1 ch·ªØ s·ªë (0-9)
            print(don_vi[hang_don_vi].capitalize())

        elif hang_chuc == 1:
            # S·ªë t·ª´ 10-19
            if hang_don_vi == 0:
                print("M∆∞·ªùi")
            elif hang_don_vi == 5:
                print("M∆∞·ªùi lƒÉm")
            else:
                print(f"M∆∞·ªùi {don_vi[hang_don_vi]}")

        else:
            # S·ªë t·ª´ 20-99
            if hang_don_vi == 0:
                print(chuc[hang_chuc].capitalize())
            elif hang_don_vi == 1:
                print(f"{chuc[hang_chuc]} m·ªët")
            elif hang_don_vi == 5:
                print(f"{chuc[hang_chuc]} lƒÉm")
            else:
                print(f"{chuc[hang_chuc]} {don_vi[hang_don_vi]}")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")

if __name__ == "__main__":
    doc_so_co_2_chu_so()
```

```python
#C√¢u 6
import math

def giai_phuong_trinh_bac_2():
    """Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c 2: ax¬≤ + bx + c = 0"""
    try:
        print("GI·∫¢I PH∆Ø∆†NG TR√åNH B·∫¨C 2: ax¬≤ + bx + c = 0")
        print("=" * 40)

        # Nh·∫≠p c√°c h·ªá s·ªë
        a = float(input("Nh·∫≠p h·ªá s·ªë a: "))
        b = float(input("Nh·∫≠p h·ªá s·ªë b: "))
        c = float(input("Nh·∫≠p h·ªá s·ªë c: "))

        print(f"\nPh∆∞∆°ng tr√¨nh: {a}x¬≤ + {b}x + {c} = 0")

        # X·ª≠ l√Ω c√°c tr∆∞·ªùng h·ª£p ƒë·∫∑c bi·ªát
        if a == 0:
            if b == 0:
                if c == 0:
                    print("Ph∆∞∆°ng tr√¨nh c√≥ v√¥ s·ªë nghi·ªám")
                else:
                    print("Ph∆∞∆°ng tr√¨nh v√¥ nghi·ªám")
            else:
                x = -c / b
                print(f"Ph∆∞∆°ng tr√¨nh c√≥ nghi·ªám duy nh·∫•t: x = {x:.2f}")
            return

        # T√≠nh delta
        delta = b**2 - 4*a*c
        print(f"Delta (Œî) = {delta:.2f}")

        # X√©t c√°c tr∆∞·ªùng h·ª£p c·ªßa delta
        if delta > 0:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print("Ph∆∞∆°ng tr√¨nh c√≥ 2 nghi·ªám ph√¢n bi·ªát:")
            print(f"x‚ÇÅ = {x1:.2f}")
            print(f"x‚ÇÇ = {x2:.2f}")

        elif delta == 0:
            x = -b / (2*a)
            print(f"Ph∆∞∆°ng tr√¨nh c√≥ nghi·ªám k√©p: x = {x:.2f}")

        else:
            # Delta < 0: ph∆∞∆°ng tr√¨nh c√≥ nghi·ªám ph·ª©c
            phan_thuc = -b / (2*a)
            phan_ao = math.sqrt(-delta) / (2*a)
            print("Ph∆∞∆°ng tr√¨nh c√≥ 2 nghi·ªám ph·ª©c:")
            print(f"x‚ÇÅ = {phan_thuc:.2f} + {phan_ao:.2f}i")
            print(f"x‚ÇÇ = {phan_thuc:.2f} - {phan_ao:.2f}i")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")
    except ZeroDivisionError:
        print("L·ªói: Kh√¥ng th·ªÉ chia cho 0!")

if __name__ == "__main__":
    giai_phuong_trinh_bac_2()
```

```python
#C√¢u 7
def tinh_tong_n_n2_n3():
    """T√≠nh t·ªïng n + n¬≤ + n¬≥"""
    try:
        # Nh·∫≠p s·ªë nguy√™n n t·ª´ b√†n ph√≠m
        n = int(input("Nh·∫≠p s·ªë nguy√™n n: "))

        # T√≠nh c√°c gi√° tr·ªã
        n1 = n
        n2 = n ** 2
        n3 = n ** 3

        # T√≠nh t·ªïng
        tong = n1 + n2 + n3

        # In k·∫øt qu·∫£
        print("\n" + "="*40)
        print(f"n = {n}")
        print(f"n¬≤ = {n2}")
        print(f"n¬≥ = {n3}")
        print(f"T·ªïng: {n} + {n2} + {n3} = {tong}")
        print("="*40)

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")

if __name__ == "__main__":
    tinh_tong_n_n2_n3()
```

```python
#C√¢u 8
def in_bang_cuu_chuong():
    """In b·∫£ng c·ª≠u ch∆∞∆°ng c·ªßa s·ªë n"""
    try:
        # Nh·∫≠p s·ªë n t·ª´ ng∆∞·ªùi d√πng
        n = int(input("Nh·∫≠p s·ªë n (1 < n < 9): "))

        # Ki·ªÉm tra ƒëi·ªÅu ki·ªán
        if n <= 1 or n >= 9:
            print("S·ªë n ph·∫£i l·ªõn h∆°n 1 v√† nh·ªè h∆°n 9!")
            return

        # In b·∫£ng c·ª≠u ch∆∞∆°ng
        print(f"\n{'='*30}")
        print(f"B·∫¢NG C·ª¨U CH∆Ø∆†NG {n}")
        print(f"{'='*30}")

        for i in range(1, 11):
            ket_qua = n * i
            print(f"{n} √ó {i:2d} = {ket_qua:2d}")

        print(f"{'='*30}")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")

if __name__ == "__main__":
    in_bang_cuu_chuong()
```

```python
#C√¢u 9
def tinh_cac_tong():
    """T√≠nh c√°c t·ªïng S, S1, S2, S3, S4"""
    try:
        # Nh·∫≠p s·ªë nguy√™n d∆∞∆°ng n
        n = int(input("Nh·∫≠p s·ªë nguy√™n d∆∞∆°ng n: "))

        # Ki·ªÉm tra ƒëi·ªÅu ki·ªán
        if n <= 0:
            print("S·ªë n ph·∫£i l√† nguy√™n d∆∞∆°ng!")
            return

        # T√≠nh c√°c t·ªïng
        S = 0      # 1 + 2 + ... + n
        S1 = 0     # 1 + 3 + ... + (2n-1)
        S2 = 0     # 2 + 4 + ... + 2n
        S3 = 0     # 1¬≤ + 2¬≤ + ... + n¬≤
        S4 = 0.0   # 1/1 + 1/2 + ... + 1/n

        # T√≠nh to√°n b·∫±ng v√≤ng l·∫∑p
        for i in range(1, n + 1):
            S += i
            S3 += i ** 2
            S4 += 1 / i

        for i in range(1, 2 * n + 1, 2):  # S·ªë l·∫ª: 1, 3, 5, ..., (2n-1)
            S1 += i

        for i in range(2, 2 * n + 1, 2):  # S·ªë ch·∫µn: 2, 4, 6, ..., 2n
            S2 += i

        # In k·∫øt qu·∫£
        print(f"\n{'='*50}")
        print(f"K·∫æT QU·∫¢ V·ªöI n = {n}")
        print(f"{'='*50}")
        print(f"S  = 1 + 2 + ... + {n} = {S}")
        print(f"S1 = 1 + 3 + ... + {2*n-1} = {S1}")
        print(f"S2 = 2 + 4 + ... + {2*n} = {S2}")
        print(f"S3 = 1¬≤ + 2¬≤ + ... + {n}¬≤ = {S3}")
        print(f"S4 = 1/1 + 1/2 + ... + 1/{n} = {S4:.6f}")
        print(f"{'='*50}")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")
    except ZeroDivisionError:
        print("L·ªói: Kh√¥ng th·ªÉ chia cho 0!")

if __name__ == "__main__":
    tinh_cac_tong()
```

```python
#C√¢u 11
import math

def tinh_giai_thua(n):
    """T√≠nh giai th·ª´a c·ªßa n"""
    if n == 0 or n == 1:
        return 1
    return n * tinh_giai_thua(n - 1)

def tinh_day_so():
    """T√≠nh d√£y s·ªë S(x,n) = x + x¬≤/2! + x¬≥/3! + ... + x‚Åø/n!"""
    try:
        # Nh·∫≠p gi√° tr·ªã x v√† n
        x = float(input("Nh·∫≠p gi√° tr·ªã x: "))
        n = int(input("Nh·∫≠p gi√° tr·ªã n (s·ªë nguy√™n d∆∞∆°ng): "))

        # Ki·ªÉm tra ƒëi·ªÅu ki·ªán
        if n <= 0:
            print("n ph·∫£i l√† s·ªë nguy√™n d∆∞∆°ng!")
            return

        S = 0.0
        print(f"\nT√≠nh S({x}, {n}) = ", end="")

        # T√≠nh t·ªïng v√† in t·ª´ng s·ªë h·∫°ng
        for i in range(1, n + 1):
            tu_so = x ** i
            mau_so = math.factorial(i)  # Ho·∫∑c c√≥ th·ªÉ d√πng tinh_giai_thua(i)
            so_hang = tu_so / mau_so
            S += so_hang

            # In s·ªë h·∫°ng
            if i == 1:
                print(f"{x}", end="")
            else:
                print(f" + ({x}^{i}/{i}!)", end="")

            # Hi·ªÉn th·ªã gi√° tr·ªã s·ªë h·∫°ng (t√πy ch·ªçn)
            # print(f" [={so_hang:.6f}]", end="")

        print(f"\n\nK·∫øt qu·∫£: S({x}, {n}) = {S:.8f}")

        # In chi ti·∫øt c√°c s·ªë h·∫°ng
        print(f"\nChi ti·∫øt c√°c s·ªë h·∫°ng:")
        print("-" * 50)
        for i in range(1, n + 1):
            tu_so = x ** i
            mau_so = math.factorial(i)
            so_hang = tu_so / mau_so
            print(f"S·ªë h·∫°ng {i}: ({x}^{i}/{i}!) = {tu_so:.2f}/{mau_so} = {so_hang:.8f}")

        print(f"\nT·ªïng S = {S:.8f}")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")

if __name__ == "__main__":
    tinh_day_so()
```

```python
#C√¢u 12
def decimal_to_binary(decimal_num):
    """Chuy·ªÉn ƒë·ªïi s·ªë th·∫≠p ph√¢n sang nh·ªã ph√¢n"""
    if decimal_num == 0:
        return "0"

    binary_str = ""
    num = abs(decimal_num)

    while num > 0:
        binary_str = str(num % 2) + binary_str
        num = num // 2

    # Th√™m d·∫•u √¢m n·∫øu c·∫ßn
    if decimal_num < 0:
        binary_str = "-" + binary_str

    return binary_str

def decimal_to_binary_builtin(decimal_num):
    """S·ª≠ d·ª•ng h√†m built-in c·ªßa Python"""
    return bin(decimal_num)[2:] if decimal_num >= 0 else "-" + bin(decimal_num)[3:]

def chuyen_doi_thap_phan_nhi_phan():
    """Ch∆∞∆°ng tr√¨nh ch√≠nh chuy·ªÉn ƒë·ªïi s·ªë th·∫≠p ph√¢n sang nh·ªã ph√¢n"""
    try:
        # Nh·∫≠p s·ªë th·∫≠p ph√¢n
        decimal_num = int(input("Nh·∫≠p s·ªë nguy√™n h·ªá th·∫≠p ph√¢n: "))

        # Chuy·ªÉn ƒë·ªïi b·∫±ng thu·∫≠t to√°n t·ª± vi·∫øt
        binary_result = decimal_to_binary(decimal_num)

        # Chuy·ªÉn ƒë·ªïi b·∫±ng h√†m built-in ƒë·ªÉ ki·ªÉm tra
        binary_builtin = decimal_to_binary_builtin(decimal_num)

        # In k·∫øt qu·∫£
        print(f"\n{'='*50}")
        print(f"S·ªë th·∫≠p ph√¢n: {decimal_num}")
        print(f"S·ªë nh·ªã ph√¢n (t·ª± vi·∫øt): {binary_result}")
        print(f"S·ªë nh·ªã ph√¢n (built-in): {binary_builtin}")
        print(f"{'='*50}")

        # Hi·ªÉn th·ªã qu√° tr√¨nh chuy·ªÉn ƒë·ªïi (n·∫øu s·ªë d∆∞∆°ng)
        if decimal_num > 0:
            print("\nüìä Qu√° tr√¨nh chuy·ªÉn ƒë·ªïi:")
            print("-" * 30)
            num = decimal_num
            step = 1
            while num > 0:
                remainder = num % 2
                quotient = num // 2
                print(f"B∆∞·ªõc {step}: {num} √∑ 2 = {quotient} d∆∞ {remainder}")
                num = quotient
                step += 1

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")

if __name__ == "__main__":
    chuyen_doi_thap_phan_nhi_phan()
```

```python
#C√¢u 13
import math

def giai_phuong_trinh_bac_2():
    """
    Gi·∫£i ph∆∞∆°ng tr√¨nh b·∫≠c 2: ax¬≤ + bx + c = 0
    """

    print("GI·∫¢I PH∆Ø∆†NG TR√åNH B·∫¨C 2: ax¬≤ + bx + c = 0")
    print("=" * 40)

    try:
        # Nh·∫≠p c√°c h·ªá s·ªë t·ª´ b√†n ph√≠m
        a = float(input("Nh·∫≠p h·ªá s·ªë a (a ‚â† 0): "))

        # Ki·ªÉm tra a kh√°c 0
        if a == 0:
            print("L·ªói: H·ªá s·ªë a ph·∫£i kh√°c 0!")
            return

        b = float(input("Nh·∫≠p h·ªá s·ªë b: "))
        c = float(input("Nh·∫≠p h·ªá s·ªë c: "))

        print(f"\nPh∆∞∆°ng tr√¨nh: {a}x¬≤ + {b}x + {c} = 0")

        # T√≠nh delta
        delta = b**2 - 4*a*c

        print(f"Delta (Œî) = {delta}")

        # X√©t c√°c tr∆∞·ªùng h·ª£p c·ªßa delta
        if delta > 0:
            # Hai nghi·ªám ph√¢n bi·ªát
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print("Ph∆∞∆°ng tr√¨nh c√≥ 2 nghi·ªám ph√¢n bi·ªát:")
            print(f"x1 = {x1}")
            print(f"x2 = {x2}")

        elif delta == 0:
            # Nghi·ªám k√©p
            x = -b / (2*a)
            print("Ph∆∞∆°ng tr√¨nh c√≥ nghi·ªám k√©p:")
            print(f"x1 = x2 = {x}")

        else:
            # V√¥ nghi·ªám (trong t·∫≠p s·ªë th·ª±c)
            print("Ph∆∞∆°ng tr√¨nh v√¥ nghi·ªám trong t·∫≠p s·ªë th·ª±c")
            # T√≠nh ph·∫ßn th·ª±c v√† ph·∫ßn ·∫£o cho nghi·ªám ph·ª©c
            phan_thuc = -b / (2*a)
            phan_ao = math.sqrt(-delta) / (2*a)
            print("Nghi·ªám ph·ª©c:")
            print(f"x1 = {phan_thuc} + {phan_ao}i")
            print(f"x2 = {phan_thuc} - {phan_ao}i")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")
    except Exception as e:
        print(f"L·ªói: {e}")

# Ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    giai_phuong_trinh_bac_2()
```

```python
#C√¢u 15
def kiem_tra_tam_giac():
    """
    Ki·ªÉm tra 3 s·ªë c√≥ th·ªÉ t·∫°o th√†nh tam gi√°c hay kh√¥ng
    """
    try:
        # Nh·∫≠p 3 s·ªë t·ª´ b√†n ph√≠m
        nhap = input("Nh·∫≠p 3 s·ªë d∆∞∆°ng a, b, c (c√°ch nhau b·∫±ng d·∫•u ph·∫©y): ")

        # T√°ch c√°c s·ªë v√† chuy·ªÉn th√†nh float
        numbers = nhap.split(',')

        # Ki·ªÉm tra ƒë·ªß 3 s·ªë
        if len(numbers) != 3:
            print("Vui l√≤ng nh·∫≠p ƒë√∫ng 3 s·ªë!")
            return

        # Chuy·ªÉn ƒë·ªïi th√†nh s·ªë v√† ki·ªÉm tra t√≠nh h·ª£p l·ªá
        a = float(numbers[0].strip())
        b = float(numbers[1].strip())
        c = float(numbers[2].strip())

        # Ki·ªÉm tra s·ªë d∆∞∆°ng
        if a <= 0 or b <= 0 or c <= 0:
            print("C√°c s·ªë ph·∫£i l√† s·ªë d∆∞∆°ng!")
            return

        # Ki·ªÉm tra ƒëi·ªÅu ki·ªán tam gi√°c
        if (a + b > c) and (a + c > b) and (b + c > a):
            print("C√≥ th·ªÉ t·∫°o th√†nh m·ªôt tam gi√°c")
        else:
            print("Kh√¥ng th·ªÉ t·∫°o th√†nh m·ªôt tam gi√°c")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")
    except Exception as e:
        print(f"L·ªói: {e}")

# Ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    kiem_tra_tam_giac()
```

```python
#C√¢u 16
def kiem_tra_ngay_thang_nam():
    """
    Ki·ªÉm tra t√≠nh h·ª£p l·ªá c·ªßa ng√†y, th√°ng, nƒÉm
    """
    try:
        # Nh·∫≠p d·ªØ li·ªáu t·ª´ b√†n ph√≠m
        print("NH·∫¨P NG√ÄY, TH√ÅNG, NƒÇM")
        print("=" * 20)

        day = int(input("Nh·∫≠p ng√†y (day): "))
        month = int(input("Nh·∫≠p th√°ng (month): "))
        year = int(input("Nh·∫≠p nƒÉm (year): "))

        # Ki·ªÉm tra t√≠nh h·ª£p l·ªá c∆° b·∫£n
        if year <= 0:
            print("NƒÉm kh√¥ng h·ª£p l·ªá! NƒÉm ph·∫£i l·ªõn h∆°n 0")
            return False

        if month < 1 or month > 12:
            print("Th√°ng kh√¥ng h·ª£p l·ªá! Th√°ng ph·∫£i t·ª´ 1 ƒë·∫øn 12")
            return False

        # Ki·ªÉm tra s·ªë ng√†y trong th√°ng
        # Th√°ng c√≥ 31 ng√†y: 1, 3, 5, 7, 8, 10, 12
        if month in [1, 3, 5, 7, 8, 10, 12]:
            max_days = 31
        # Th√°ng c√≥ 30 ng√†y: 4, 6, 9, 11
        elif month in [4, 6, 9, 11]:
            max_days = 30
        # Th√°ng 2 - ki·ªÉm tra nƒÉm nhu·∫≠n
        else:  # month == 2
            if (year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
                max_days = 29  # NƒÉm nhu·∫≠n
            else:
                max_days = 28  # NƒÉm kh√¥ng nhu·∫≠n

        # Ki·ªÉm tra ng√†y
        if day < 1 or day > max_days:
            print(f"Ng√†y kh√¥ng h·ª£p l·ªá! Th√°ng {month} nƒÉm {year} c√≥ t·ªëi ƒëa {max_days} ng√†y")
            return False

        # N·∫øu t·∫•t c·∫£ ƒë·ªÅu h·ª£p l·ªá
        print(f"Ng√†y {day}/{month}/{year} l√† h·ª£p l·ªá!")
        return True

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")
        return False
    except Exception as e:
        print(f"L·ªói kh√¥ng x√°c ƒë·ªãnh: {e}")
        return False

# Ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    kiem_tra_ngay_thang_nam()
```

```python
#C√¢u 17
def tinh_cac_tong():
    """
    T√≠nh c√°c t·ªïng theo y√™u c·∫ßu
    """
    try:
        # Nh·∫≠p s·ªë t·ª± nhi√™n n
        n = int(input("Nh·∫≠p s·ªë t·ª± nhi√™n n: "))

        if n < 0:
            print("Vui l√≤ng nh·∫≠p s·ªë t·ª± nhi√™n (n ‚â• 0)")
            return

        print(f"\nK·∫øt qu·∫£ v·ªõi n = {n}:")
        print("=" * 30)

        # a) T√≠nh t·ªïng 1 + 2 + ... + 2n
        tong_a = 0
        for i in range(1, 2*n + 1):
            tong_a += i

        print(f"a) T·ªïng 1 + 2 + ... + 2n = {tong_a}")
        print(f"   C√¥ng th·ª©c: (2n √ó (2n + 1)) / 2 = {(2*n * (2*n + 1)) // 2}")

        # b) T√≠nh t·ªïng c√°c s·ªë t·ª± nhi√™n < n v√† l√† s·ªë l·∫ª
        tong_b = 0
        so_le = []
        for i in range(1, n):
            if i % 2 != 0:  # S·ªë l·∫ª
                tong_b += i
                so_le.append(i)

        print(f"b) T·ªïng c√°c s·ªë t·ª± nhi√™n < {n} v√† l√† s·ªë l·∫ª = {tong_b}")
        if so_le:
            print(f"   C√°c s·ªë l·∫ª: {so_le}")
        else:
            print("   Kh√¥ng c√≥ s·ªë l·∫ª n√†o")

        # c) T√≠nh t·ªïng c√°c s·ªë t·ª± nhi√™n < n v√† l√† s·ªë ch·∫µn
        tong_c = 0
        so_chan = []
        for i in range(0, n):
            if i % 2 == 0:  # S·ªë ch·∫µn
                tong_c += i
                so_chan.append(i)

        print(f"c) T·ªïng c√°c s·ªë t·ª± nhi√™n < {n} v√† l√† s·ªë ch·∫µn = {tong_c}")
        if so_chan:
            print(f"   C√°c s·ªë ch·∫µn: {so_chan}")
        else:
            print("   Kh√¥ng c√≥ s·ªë ch·∫µn n√†o")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë nguy√™n h·ª£p l·ªá!")
    except Exception as e:
        print(f"L·ªói kh√¥ng x√°c ƒë·ªãnh: {e}")

# Ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    tinh_cac_tong()
```

```python
#C√¢u 18
import math

def dem_uoc_nguyen_to_don_gian(n):
    """
    ƒê·∫øm s·ªë ∆∞·ªõc nguy√™n t·ªë kh√°c nhau (phi√™n b·∫£n ƒë∆°n gi·∫£n)
    """
    if n < 2:
        return 0

    count = 0
    temp = n

    # Ki·ªÉm tra ∆∞·ªõc 2
    if temp % 2 == 0:
        count += 1
        while temp % 2 == 0:
            temp //= 2

    # Ki·ªÉm tra c√°c ∆∞·ªõc l·∫ª
    i = 3
    while i <= temp:
        if temp % i == 0:
            count += 1
            while temp % i == 0:
                temp //= i
        i += 2

    return count

def tong_uoc_thuc_su_don_gian(n):
    """
    T√≠nh t·ªïng ∆∞·ªõc th·ª±c s·ª± (phi√™n b·∫£n ƒë∆°n gi·∫£n)
    """
    if n <= 1:
        return 0

    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

def tinh_cac_gia_tri():
    """
    Ch∆∞∆°ng tr√¨nh t√≠nh c√°c gi√° tr·ªã (phi√™n b·∫£n ƒë∆°n gi·∫£n)
    """
    try:
        n = int(input("Nh·∫≠p s·ªë t·ª± nhi√™n n: "))

        if n < 0:
            print("Vui l√≤ng nh·∫≠p s·ªë t·ª± nhi√™n!")
            return

        print(f"\nK·∫øt qu·∫£ v·ªõi n = {n}:")

        # a) S·ªë ∆∞·ªõc nguy√™n t·ªë
        so_uoc_nguyen_to = dem_uoc_nguyen_to_don_gian(n)
        print(f"a) S·ªë ∆∞·ªõc s·ªë nguy√™n t·ªë kh√°c nhau: {so_uoc_nguyen_to}")

        # b) T·ªïng ∆∞·ªõc th·ª±c s·ª±
        tong_uoc = tong_uoc_thuc_su_don_gian(n)
        print(f"b) T·ªïng c√°c ∆∞·ªõc s·ªë th·ª±c s·ª±: {tong_uoc}")

        # c) T·ªïng b√¨nh ph∆∞∆°ng
        if n >= 1:
            tong_bp = n * (n + 1) * (2*n + 1) // 6
            print(f"c) T·ªïng 1¬≤ + 2¬≤ + ... + {n}¬≤ = {tong_bp}")
        else:
            print("c) T·ªïng b√¨nh ph∆∞∆°ng = 0")

    except ValueError:
        print("L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")

# Ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    tinh_cac_gia_tri()
```

```python
#C√¢u 19
# a) T√≠nh t·ªïng 1+2+...+n
def tong_day_so(n):
    """In ra t·ªïng 1+2+...+n"""
    if n < 0:
        print("L·ªói: n ph·∫£i l√† s·ªë t·ª± nhi√™n")
        return
    tong = n * (n + 1) // 2
    print(f"1+2+...+{n} = {tong}")

# b) Ki·ªÉm tra nguy√™n t·ªë c√πng nhau
def kiem_tra_nguyen_to_cung_nhau(m, n):
    """Ki·ªÉm tra v√† in th√¥ng b√°o"""
    if m <= 0 or n <= 0:
        print("L·ªói: C√°c s·ªë ph·∫£i d∆∞∆°ng")
        return

    # T√¨m UCLN
    a, b = m, n
    while b != 0:
        a, b = b, a % b

    if a == 1:
        print(f"{m} v√† {n} l√† nguy√™n t·ªë c√πng nhau")
    else:
        print(f"{m} v√† {n} kh√¥ng l√† nguy√™n t·ªë c√πng nhau")

# c) Chuy·ªÉn ƒë·ªïi gi√¢y
def chuyen_doi_giay(s):
    """Chuy·ªÉn ƒë·ªïi v√† in ra gi·ªù:ph√∫t:gi√¢y"""
    if s < 0:
        print("L·ªói: S·ªë gi√¢y ph·∫£i kh√¥ng √¢m")
        return

    gio = s // 3600
    phut = (s % 3600) // 60
    giay = s % 60
    print(f"{s} gi√¢y = {gio}:{phut:02d}:{giay:02d}")

# d) T√≠nh tr·ªã tuy·ªát ƒë·ªëi |a-b|
def tri_tuyet_doi_hieu(a, b):
    """T√≠nh v√† tr·∫£ v·ªÅ |a-b|"""
    result = abs(a - b)
    print(f"|{a} - {b}| = {result}")
    return result

if __name__ == "__main__":
    # Test c√°c h√†m
    tong_day_so(10)
    kiem_tra_nguyen_to_cung_nhau(8, 9)
    chuyen_doi_giay(3665)
    ket_qua = tri_tuyet_doi_hieu(5.5, 3.2)
```

```python
#C√¢u 20
import math

def is_prime(num):
    """
    Ki·ªÉm tra s·ªë nguy√™n t·ªë
    """
    if num < 2:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False

    # Ki·ªÉm tra t·ª´ 3 ƒë·∫øn cƒÉn b·∫≠c 2 c·ªßa num
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False
    return True

def f(m, n):
    """
    T√¨m s·ªë nguy√™n t·ªë l·ªõn nh·∫•t ƒë·ªìng th·ªùi l√† ∆∞·ªõc chung c·ªßa m v√† n
    Tr·∫£ v·ªÅ 0 n·∫øu kh√¥ng t·ªìn t·∫°i
    """
    # Ki·ªÉm tra ƒëi·ªÅu ki·ªán ƒë·∫ßu v√†o
    if m <= 0 or n <= 0:
        return 0

    # T√¨m ∆∞·ªõc chung l·ªõn nh·∫•t
    def gcd(a, b):
        while b != 0:
            a, b = b, a % b
        return a

    ucln = gcd(m, n)

    # N·∫øu UCLN = 1, kh√¥ng c√≥ ∆∞·ªõc nguy√™n t·ªë chung
    if ucln == 1:
        return 0

    # T√¨m c√°c ∆∞·ªõc nguy√™n t·ªë c·ªßa UCLN
    prime_factors = []
    temp = ucln

    # Ki·ªÉm tra ∆∞·ªõc nguy√™n t·ªë 2
    if temp % 2 == 0:
        prime_factors.append(2)
        while temp % 2 == 0:
            temp //= 2

    # Ki·ªÉm tra c√°c ∆∞·ªõc nguy√™n t·ªë l·∫ª
    i = 3
    while i <= temp:
        if temp % i == 0 and is_prime(i):
            prime_factors.append(i)
            while temp % i == 0:
                temp //= i
        i += 2

    # N·∫øu c√≤n s·ªë nguy√™n t·ªë l·ªõn h∆°n 2
    if temp > 2 and is_prime(temp):
        prime_factors.append(temp)

    # Tr·∫£ v·ªÅ s·ªë nguy√™n t·ªë l·ªõn nh·∫•t
    if prime_factors:
        return max(prime_factors)
    else:
        return 0

# Ch∆∞∆°ng tr√¨nh ch√≠nh
def main():
    """
    Ch∆∞∆°ng tr√¨nh ch√≠nh nh·∫≠p li·ªáu t·ª´ ng∆∞·ªùi d√πng
    """
    print("T√åM S·ªê NGUY√äN T·ªê L·ªöN NH·∫§T L√Ä ∆Ø·ªöC CHUNG")
    print("=" * 50)

    while True:
        try:
            print("\nNh·∫≠p hai s·ªë t·ª± nhi√™n (ho·∫∑c 'q' ƒë·ªÉ tho√°t):")
            m_input = input("Nh·∫≠p m: ").strip()
            if m_input.lower() == 'q':
                break

            n_input = input("Nh·∫≠p n: ").strip()
            if n_input.lower() == 'q':
                break

            m = int(m_input)
            n = int(n_input)

            if m <= 0 or n <= 0:
                print(" Vui l√≤ng nh·∫≠p s·ªë t·ª± nhi√™n d∆∞∆°ng!")
                continue

            result = f(m, n)

            if result == 0:
                print(f" Kh√¥ng c√≥ s·ªë nguy√™n t·ªë n√†o l√† ∆∞·ªõc chung c·ªßa {m} v√† {n}")
            else:
                print(f" S·ªë nguy√™n t·ªë l·ªõn nh·∫•t l√† ∆∞·ªõc chung c·ªßa {m} v√† {n} l√†: {result}")

        except ValueError:
            print(" L·ªói: Vui l√≤ng nh·∫≠p s·ªë h·ª£p l·ªá!")
        except Exception as e:
            print(f" L·ªói kh√¥ng x√°c ƒë·ªãnh: {e}")

# Ch·∫°y ch∆∞∆°ng tr√¨nh
if __name__ == "__main__":
    # Ch·∫°y ch∆∞∆°ng tr√¨nh ch√≠nh
    main()
```
