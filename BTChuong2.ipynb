```python
#Câu 1
def main():
    try:
        # Nhập số tiền gốc từ người dùng
        X = float(input("Nhập số tiền gốc (X): "))

        # Kiểm tra số tiền hợp lệ
        if X <= 0:
            print("Số tiền phải lớn hơn 0!")
            return

        # Thiết lập các tham số
        lai_suat_thang = 0.006  # 0.6% = 0.006
        so_thang = 18
        ky_han = 6  # Cứ 6 tháng cộng lãi vào gốc một lần

        # Tính số kỳ hạn trong 18 tháng
        so_ky_han = so_thang // ky_han

        # Tính tổng tiền sau mỗi kỳ hạn
        tong_tien = X

        for ky in range(so_ky_han):
            # Tính lãi suất cho 6 tháng
            lai_suat_6_thang = (1 + lai_suat_thang) ** ky_han - 1

            # Tính tiền lãi cho kỳ này
            tien_lai = tong_tien * lai_suat_6_thang

            # Cộng lãi vào gốc
            tong_tien += tien_lai

            print(f"Sau {ky_han * (ky + 1)} tháng: {tong_tien:,.0f} VND")

        print("\n" + "="*50)
        print(f"Số tiền gốc ban đầu: {X:,.0f} VND")
        print(f"Lãi suất tháng: {lai_suat_thang * 100}%")
        print(f"Sau {so_thang} tháng, tổng số tiền là: {tong_tien:,.0f} VND")

    except ValueError:
        print("Vui lòng nhập số tiền hợp lệ!")

if __name__ == "__main__":
    main()
```

```python
#Câu 5
def doc_so_co_2_chu_so():
    """Chương trình đọc số có tối đa 2 chữ số"""
    try:
        # Nhập số từ người dùng
        n = int(input("Nhập một số (0-99): "))

        # Kiểm tra số hợp lệ
        if n < 0 or n > 99:
            print("Số phải từ 0 đến 99!")
            return

        # Danh sách cách đọc các số
        don_vi = ["không", "một", "hai", "ba", "bốn", "năm", "sáu", "bảy", "tám", "chín"]
        chuc = ["", "mười", "hai mươi", "ba mươi", "bốn mươi", "năm mươi",
                "sáu mươi", "bảy mươi", "tám mươi", "chín mươi"]

        # Xử lý các trường hợp đặc biệt
        if n == 0:
            print("Không")
            return

        # Tách chữ số hàng chục và hàng đơn vị
        hang_chuc = n // 10
        hang_don_vi = n % 10

        # Đọc số theo các trường hợp
        if hang_chuc == 0:
            # Số có 1 chữ số (0-9)
            print(don_vi[hang_don_vi].capitalize())

        elif hang_chuc == 1:
            # Số từ 10-19
            if hang_don_vi == 0:
                print("Mười")
            elif hang_don_vi == 5:
                print("Mười lăm")
            else:
                print(f"Mười {don_vi[hang_don_vi]}")

        else:
            # Số từ 20-99
            if hang_don_vi == 0:
                print(chuc[hang_chuc].capitalize())
            elif hang_don_vi == 1:
                print(f"{chuc[hang_chuc]} mốt")
            elif hang_don_vi == 5:
                print(f"{chuc[hang_chuc]} lăm")
            else:
                print(f"{chuc[hang_chuc]} {don_vi[hang_don_vi]}")

    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên hợp lệ!")

if __name__ == "__main__":
    doc_so_co_2_chu_so()
```

```python
#Câu 6
import math

def giai_phuong_trinh_bac_2():
    """Giải phương trình bậc 2: ax² + bx + c = 0"""
    try:
        print("GIẢI PHƯƠNG TRÌNH BẬC 2: ax² + bx + c = 0")
        print("=" * 40)

        # Nhập các hệ số
        a = float(input("Nhập hệ số a: "))
        b = float(input("Nhập hệ số b: "))
        c = float(input("Nhập hệ số c: "))

        print(f"\nPhương trình: {a}x² + {b}x + {c} = 0")

        # Xử lý các trường hợp đặc biệt
        if a == 0:
            if b == 0:
                if c == 0:
                    print("Phương trình có vô số nghiệm")
                else:
                    print("Phương trình vô nghiệm")
            else:
                x = -c / b
                print(f"Phương trình có nghiệm duy nhất: x = {x:.2f}")
            return

        # Tính delta
        delta = b**2 - 4*a*c
        print(f"Delta (Δ) = {delta:.2f}")

        # Xét các trường hợp của delta
        if delta > 0:
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print("Phương trình có 2 nghiệm phân biệt:")
            print(f"x₁ = {x1:.2f}")
            print(f"x₂ = {x2:.2f}")

        elif delta == 0:
            x = -b / (2*a)
            print(f"Phương trình có nghiệm kép: x = {x:.2f}")

        else:
            # Delta < 0: phương trình có nghiệm phức
            phan_thuc = -b / (2*a)
            phan_ao = math.sqrt(-delta) / (2*a)
            print("Phương trình có 2 nghiệm phức:")
            print(f"x₁ = {phan_thuc:.2f} + {phan_ao:.2f}i")
            print(f"x₂ = {phan_thuc:.2f} - {phan_ao:.2f}i")

    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ!")
    except ZeroDivisionError:
        print("Lỗi: Không thể chia cho 0!")

if __name__ == "__main__":
    giai_phuong_trinh_bac_2()
```

```python
#Câu 9
def tinh_cac_tong():
    """Tính các tổng S, S1, S2, S3, S4"""
    try:
        # Nhập số nguyên dương n
        n = int(input("Nhập số nguyên dương n: "))

        # Kiểm tra điều kiện
        if n <= 0:
            print("Số n phải là nguyên dương!")
            return

        # Tính các tổng
        S = 0      # 1 + 2 + ... + n
        S1 = 0     # 1 + 3 + ... + (2n-1)
        S2 = 0     # 2 + 4 + ... + 2n
        S3 = 0     # 1² + 2² + ... + n²
        S4 = 0.0   # 1/1 + 1/2 + ... + 1/n

        # Tính toán bằng vòng lặp
        for i in range(1, n + 1):
            S += i
            S3 += i ** 2
            S4 += 1 / i

        for i in range(1, 2 * n + 1, 2):  # Số lẻ: 1, 3, 5, ..., (2n-1)
            S1 += i

        for i in range(2, 2 * n + 1, 2):  # Số chẵn: 2, 4, 6, ..., 2n
            S2 += i

        # In kết quả
        print(f"\n{'='*50}")
        print(f"KẾT QUẢ VỚI n = {n}")
        print(f"{'='*50}")
        print(f"S  = 1 + 2 + ... + {n} = {S}")
        print(f"S1 = 1 + 3 + ... + {2*n-1} = {S1}")
        print(f"S2 = 2 + 4 + ... + {2*n} = {S2}")
        print(f"S3 = 1² + 2² + ... + {n}² = {S3}")
        print(f"S4 = 1/1 + 1/2 + ... + 1/{n} = {S4:.6f}")
        print(f"{'='*50}")

    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên hợp lệ!")
    except ZeroDivisionError:
        print("Lỗi: Không thể chia cho 0!")

if __name__ == "__main__":
    tinh_cac_tong()
```

```python
#Câu 11
import math

def tinh_giai_thua(n):
    """Tính giai thừa của n"""
    if n == 0 or n == 1:
        return 1
    return n * tinh_giai_thua(n - 1)

def tinh_day_so():
    """Tính dãy số S(x,n) = x + x²/2! + x³/3! + ... + xⁿ/n!"""
    try:
        # Nhập giá trị x và n
        x = float(input("Nhập giá trị x: "))
        n = int(input("Nhập giá trị n (số nguyên dương): "))

        # Kiểm tra điều kiện
        if n <= 0:
            print("n phải là số nguyên dương!")
            return

        S = 0.0
        print(f"\nTính S({x}, {n}) = ", end="")

        # Tính tổng và in từng số hạng
        for i in range(1, n + 1):
            tu_so = x ** i
            mau_so = math.factorial(i)  # Hoặc có thể dùng tinh_giai_thua(i)
            so_hang = tu_so / mau_so
            S += so_hang

            # In số hạng
            if i == 1:
                print(f"{x}", end="")
            else:
                print(f" + ({x}^{i}/{i}!)", end="")

            # Hiển thị giá trị số hạng (tùy chọn)
            # print(f" [={so_hang:.6f}]", end="")

        print(f"\n\nKết quả: S({x}, {n}) = {S:.8f}")

        # In chi tiết các số hạng
        print(f"\nChi tiết các số hạng:")
        print("-" * 50)
        for i in range(1, n + 1):
            tu_so = x ** i
            mau_so = math.factorial(i)
            so_hang = tu_so / mau_so
            print(f"Số hạng {i}: ({x}^{i}/{i}!) = {tu_so:.2f}/{mau_so} = {so_hang:.8f}")

        print(f"\nTổng S = {S:.8f}")

    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ!")

if __name__ == "__main__":
    tinh_day_so()
```

```python
#Câu 12
def decimal_to_binary(decimal_num):
    """Chuyển đổi số thập phân sang nhị phân"""
    if decimal_num == 0:
        return "0"

    binary_str = ""
    num = abs(decimal_num)

    while num > 0:
        binary_str = str(num % 2) + binary_str
        num = num // 2

    # Thêm dấu âm nếu cần
    if decimal_num < 0:
        binary_str = "-" + binary_str

    return binary_str

def decimal_to_binary_builtin(decimal_num):
    """Sử dụng hàm built-in của Python"""
    return bin(decimal_num)[2:] if decimal_num >= 0 else "-" + bin(decimal_num)[3:]

def chuyen_doi_thap_phan_nhi_phan():
    """Chương trình chính chuyển đổi số thập phân sang nhị phân"""
    try:
        # Nhập số thập phân
        decimal_num = int(input("Nhập số nguyên hệ thập phân: "))

        # Chuyển đổi bằng thuật toán tự viết
        binary_result = decimal_to_binary(decimal_num)

        # Chuyển đổi bằng hàm built-in để kiểm tra
        binary_builtin = decimal_to_binary_builtin(decimal_num)

        # In kết quả
        print(f"\n{'='*50}")
        print(f"Số thập phân: {decimal_num}")
        print(f"Số nhị phân (tự viết): {binary_result}")
        print(f"Số nhị phân (built-in): {binary_builtin}")
        print(f"{'='*50}")

        # Hiển thị quá trình chuyển đổi (nếu số dương)
        if decimal_num > 0:
            print("\n📊 Quá trình chuyển đổi:")
            print("-" * 30)
            num = decimal_num
            step = 1
            while num > 0:
                remainder = num % 2
                quotient = num // 2
                print(f"Bước {step}: {num} ÷ 2 = {quotient} dư {remainder}")
                num = quotient
                step += 1

    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên hợp lệ!")

if __name__ == "__main__":
    chuyen_doi_thap_phan_nhi_phan()
```

```python
#Câu 13
import math

def giai_phuong_trinh_bac_2():
    """
    Giải phương trình bậc 2: ax² + bx + c = 0
    """

    print("GIẢI PHƯƠNG TRÌNH BẬC 2: ax² + bx + c = 0")
    print("=" * 40)

    try:
        # Nhập các hệ số từ bàn phím
        a = float(input("Nhập hệ số a (a ≠ 0): "))

        # Kiểm tra a khác 0
        if a == 0:
            print("Lỗi: Hệ số a phải khác 0!")
            return

        b = float(input("Nhập hệ số b: "))
        c = float(input("Nhập hệ số c: "))

        print(f"\nPhương trình: {a}x² + {b}x + {c} = 0")

        # Tính delta
        delta = b**2 - 4*a*c

        print(f"Delta (Δ) = {delta}")

        # Xét các trường hợp của delta
        if delta > 0:
            # Hai nghiệm phân biệt
            x1 = (-b + math.sqrt(delta)) / (2*a)
            x2 = (-b - math.sqrt(delta)) / (2*a)
            print("Phương trình có 2 nghiệm phân biệt:")
            print(f"x1 = {x1}")
            print(f"x2 = {x2}")

        elif delta == 0:
            # Nghiệm kép
            x = -b / (2*a)
            print("Phương trình có nghiệm kép:")
            print(f"x1 = x2 = {x}")

        else:
            # Vô nghiệm (trong tập số thực)
            print("Phương trình vô nghiệm trong tập số thực")
            # Tính phần thực và phần ảo cho nghiệm phức
            phan_thuc = -b / (2*a)
            phan_ao = math.sqrt(-delta) / (2*a)
            print("Nghiệm phức:")
            print(f"x1 = {phan_thuc} + {phan_ao}i")
            print(f"x2 = {phan_thuc} - {phan_ao}i")

    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ!")
    except Exception as e:
        print(f"Lỗi: {e}")

# Chạy chương trình
if __name__ == "__main__":
    giai_phuong_trinh_bac_2()
```

```python
#Câu 15
def kiem_tra_tam_giac():
    """
    Kiểm tra 3 số có thể tạo thành tam giác hay không
    """
    try:
        # Nhập 3 số từ bàn phím
        nhap = input("Nhập 3 số dương a, b, c (cách nhau bằng dấu phẩy): ")

        # Tách các số và chuyển thành float
        numbers = nhap.split(',')

        # Kiểm tra đủ 3 số
        if len(numbers) != 3:
            print("Vui lòng nhập đúng 3 số!")
            return

        # Chuyển đổi thành số và kiểm tra tính hợp lệ
        a = float(numbers[0].strip())
        b = float(numbers[1].strip())
        c = float(numbers[2].strip())

        # Kiểm tra số dương
        if a <= 0 or b <= 0 or c <= 0:
            print("Các số phải là số dương!")
            return

        # Kiểm tra điều kiện tam giác
        if (a + b > c) and (a + c > b) and (b + c > a):
            print("Có thể tạo thành một tam giác")
        else:
            print("Không thể tạo thành một tam giác")

    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ!")
    except Exception as e:
        print(f"Lỗi: {e}")

# Chạy chương trình
if __name__ == "__main__":
    kiem_tra_tam_giac()
```

```python
#Câu 16
def kiem_tra_ngay_thang_nam():
    """
    Kiểm tra tính hợp lệ của ngày, tháng, năm
    """
    try:
        # Nhập dữ liệu từ bàn phím
        print("NHẬP NGÀY, THÁNG, NĂM")
        print("=" * 20)

        day = int(input("Nhập ngày (day): "))
        month = int(input("Nhập tháng (month): "))
        year = int(input("Nhập năm (year): "))

        # Kiểm tra tính hợp lệ cơ bản
        if year <= 0:
            print("Năm không hợp lệ! Năm phải lớn hơn 0")
            return False

        if month < 1 or month > 12:
            print("Tháng không hợp lệ! Tháng phải từ 1 đến 12")
            return False

        # Kiểm tra số ngày trong tháng
        # Tháng có 31 ngày: 1, 3, 5, 7, 8, 10, 12
        if month in [1, 3, 5, 7, 8, 10, 12]:
            max_days = 31
        # Tháng có 30 ngày: 4, 6, 9, 11
        elif month in [4, 6, 9, 11]:
            max_days = 30
        # Tháng 2 - kiểm tra năm nhuận
        else:  # month == 2
            if (year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
                max_days = 29  # Năm nhuận
            else:
                max_days = 28  # Năm không nhuận

        # Kiểm tra ngày
        if day < 1 or day > max_days:
            print(f"Ngày không hợp lệ! Tháng {month} năm {year} có tối đa {max_days} ngày")
            return False

        # Nếu tất cả đều hợp lệ
        print(f"Ngày {day}/{month}/{year} là hợp lệ!")
        return True

    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên hợp lệ!")
        return False
    except Exception as e:
        print(f"Lỗi không xác định: {e}")
        return False

# Chạy chương trình
if __name__ == "__main__":
    kiem_tra_ngay_thang_nam()
```

```python
#Câu 17
def tinh_cac_tong():
    """
    Tính các tổng theo yêu cầu
    """
    try:
        # Nhập số tự nhiên n
        n = int(input("Nhập số tự nhiên n: "))

        if n < 0:
            print("Vui lòng nhập số tự nhiên (n ≥ 0)")
            return

        print(f"\nKết quả với n = {n}:")
        print("=" * 30)

        # a) Tính tổng 1 + 2 + ... + 2n
        tong_a = 0
        for i in range(1, 2*n + 1):
            tong_a += i

        print(f"a) Tổng 1 + 2 + ... + 2n = {tong_a}")
        print(f"   Công thức: (2n × (2n + 1)) / 2 = {(2*n * (2*n + 1)) // 2}")

        # b) Tính tổng các số tự nhiên < n và là số lẻ
        tong_b = 0
        so_le = []
        for i in range(1, n):
            if i % 2 != 0:  # Số lẻ
                tong_b += i
                so_le.append(i)

        print(f"b) Tổng các số tự nhiên < {n} và là số lẻ = {tong_b}")
        if so_le:
            print(f"   Các số lẻ: {so_le}")
        else:
            print("   Không có số lẻ nào")

        # c) Tính tổng các số tự nhiên < n và là số chẵn
        tong_c = 0
        so_chan = []
        for i in range(0, n):
            if i % 2 == 0:  # Số chẵn
                tong_c += i
                so_chan.append(i)

        print(f"c) Tổng các số tự nhiên < {n} và là số chẵn = {tong_c}")
        if so_chan:
            print(f"   Các số chẵn: {so_chan}")
        else:
            print("   Không có số chẵn nào")

    except ValueError:
        print("Lỗi: Vui lòng nhập số nguyên hợp lệ!")
    except Exception as e:
        print(f"Lỗi không xác định: {e}")

# Chạy chương trình
if __name__ == "__main__":
    tinh_cac_tong()
```

```python
#Câu 18
import math

def dem_uoc_nguyen_to_don_gian(n):
    """
    Đếm số ước nguyên tố khác nhau (phiên bản đơn giản)
    """
    if n < 2:
        return 0

    count = 0
    temp = n

    # Kiểm tra ước 2
    if temp % 2 == 0:
        count += 1
        while temp % 2 == 0:
            temp //= 2

    # Kiểm tra các ước lẻ
    i = 3
    while i <= temp:
        if temp % i == 0:
            count += 1
            while temp % i == 0:
                temp //= i
        i += 2

    return count

def tong_uoc_thuc_su_don_gian(n):
    """
    Tính tổng ước thực sự (phiên bản đơn giản)
    """
    if n <= 1:
        return 0

    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

def tinh_cac_gia_tri():
    """
    Chương trình tính các giá trị (phiên bản đơn giản)
    """
    try:
        n = int(input("Nhập số tự nhiên n: "))

        if n < 0:
            print("Vui lòng nhập số tự nhiên!")
            return

        print(f"\nKết quả với n = {n}:")

        # a) Số ước nguyên tố
        so_uoc_nguyen_to = dem_uoc_nguyen_to_don_gian(n)
        print(f"a) Số ước số nguyên tố khác nhau: {so_uoc_nguyen_to}")

        # b) Tổng ước thực sự
        tong_uoc = tong_uoc_thuc_su_don_gian(n)
        print(f"b) Tổng các ước số thực sự: {tong_uoc}")

        # c) Tổng bình phương
        if n >= 1:
            tong_bp = n * (n + 1) * (2*n + 1) // 6
            print(f"c) Tổng 1² + 2² + ... + {n}² = {tong_bp}")
        else:
            print("c) Tổng bình phương = 0")

    except ValueError:
        print("Lỗi: Vui lòng nhập số hợp lệ!")

# Chạy chương trình
if __name__ == "__main__":
    tinh_cac_gia_tri()
```

```python
#Câu 19
# a) Tính tổng 1+2+...+n
def tong_day_so(n):
    """In ra tổng 1+2+...+n"""
    if n < 0:
        print("Lỗi: n phải là số tự nhiên")
        return
    tong = n * (n + 1) // 2
    print(f"1+2+...+{n} = {tong}")

# b) Kiểm tra nguyên tố cùng nhau
def kiem_tra_nguyen_to_cung_nhau(m, n):
    """Kiểm tra và in thông báo"""
    if m <= 0 or n <= 0:
        print("Lỗi: Các số phải dương")
        return

    # Tìm UCLN
    a, b = m, n
    while b != 0:
        a, b = b, a % b

    if a == 1:
        print(f"{m} và {n} là nguyên tố cùng nhau")
    else:
        print(f"{m} và {n} không là nguyên tố cùng nhau")

# c) Chuyển đổi giây
def chuyen_doi_giay(s):
    """Chuyển đổi và in ra giờ:phút:giây"""
    if s < 0:
        print("Lỗi: Số giây phải không âm")
        return

    gio = s // 3600
    phut = (s % 3600) // 60
    giay = s % 60
    print(f"{s} giây = {gio}:{phut:02d}:{giay:02d}")

# d) Tính trị tuyệt đối |a-b|
def tri_tuyet_doi_hieu(a, b):
    """Tính và trả về |a-b|"""
    result = abs(a - b)
    print(f"|{a} - {b}| = {result}")
    return result

if __name__ == "__main__":
    # Test các hàm
    tong_day_so(10)
    kiem_tra_nguyen_to_cung_nhau(8, 9)
    chuyen_doi_giay(3665)
    ket_qua = tri_tuyet_doi_hieu(5.5, 3.2)
```

```python
#Câu 20
import math

def is_prime(num):
    """
    Kiểm tra số nguyên tố
    """
    if num < 2:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False

    # Kiểm tra từ 3 đến căn bậc 2 của num
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False
    return True

def f(m, n):
    """
    Tìm số nguyên tố lớn nhất đồng thời là ước chung của m và n
    Trả về 0 nếu không tồn tại
    """
    # Kiểm tra điều kiện đầu vào
    if m <= 0 or n <= 0:
        return 0

    # Tìm ước chung lớn nhất
    def gcd(a, b):
        while b != 0:
            a, b = b, a % b
        return a

    ucln = gcd(m, n)

    # Nếu UCLN = 1, không có ước nguyên tố chung
    if ucln == 1:
        return 0

    # Tìm các ước nguyên tố của UCLN
    prime_factors = []
    temp = ucln

    # Kiểm tra ước nguyên tố 2
    if temp % 2 == 0:
        prime_factors.append(2)
        while temp % 2 == 0:
            temp //= 2

    # Kiểm tra các ước nguyên tố lẻ
    i = 3
    while i <= temp:
        if temp % i == 0 and is_prime(i):
            prime_factors.append(i)
            while temp % i == 0:
                temp //= i
        i += 2

    # Nếu còn số nguyên tố lớn hơn 2
    if temp > 2 and is_prime(temp):
        prime_factors.append(temp)

    # Trả về số nguyên tố lớn nhất
    if prime_factors:
        return max(prime_factors)
    else:
        return 0

# Chương trình chính
def main():
    """
    Chương trình chính nhập liệu từ người dùng
    """
    print("TÌM SỐ NGUYÊN TỐ LỚN NHẤT LÀ ƯỚC CHUNG")
    print("=" * 50)

    while True:
        try:
            print("\nNhập hai số tự nhiên (hoặc 'q' để thoát):")
            m_input = input("Nhập m: ").strip()
            if m_input.lower() == 'q':
                break

            n_input = input("Nhập n: ").strip()
            if n_input.lower() == 'q':
                break

            m = int(m_input)
            n = int(n_input)

            if m <= 0 or n <= 0:
                print(" Vui lòng nhập số tự nhiên dương!")
                continue

            result = f(m, n)

            if result == 0:
                print(f" Không có số nguyên tố nào là ước chung của {m} và {n}")
            else:
                print(f" Số nguyên tố lớn nhất là ước chung của {m} và {n} là: {result}")

        except ValueError:
            print(" Lỗi: Vui lòng nhập số hợp lệ!")
        except Exception as e:
            print(f" Lỗi không xác định: {e}")

# Chạy chương trình
if __name__ == "__main__":
    # Chạy chương trình chính
    main()
```
