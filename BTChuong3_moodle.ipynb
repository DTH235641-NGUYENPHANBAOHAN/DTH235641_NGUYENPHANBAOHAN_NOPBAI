```python
#Câu 7
def la_nam_nhuan(nam):
    return (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0)

def so_ngay_trong_thang(thang, nam):
    if thang in [1, 3, 5, 7, 8, 10, 12]:
        return 31
    elif thang in [4, 6, 9, 11]:
        return 30
    elif thang == 2:
        return 29 if la_nam_nhuan(nam) else 28
    return 0

# Nhập dữ liệu
d, m, y = map(int, input("Nhap ngay thang nam (d m y): ").split())

max_day = so_ngay_trong_thang(m, y)

if d < max_day:
    d += 1
else:
    d = 1
    if m == 12:
        m = 1
        y += 1
    else:
        m += 1

print(f"Ngay ke tiep la: {d}/{m}/{y}")
```

```python
#Câu 18
def hinh1(n):
    # Hình vuông rỗng
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()


def hinh2(n):
    # Tam giác vuông cân (từ 1 đến n sao)
    for i in range(1, n+1):
        for j in range(i):
            print("*", end=" ")
        print()


def hinh3(n):
    # Hình chữ X
    for i in range(n):
        for j in range(n):
            if j == i or j == n - i - 1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()


# ==============================
# Chạy thử
n = int(input("Nhập chiều cao n: "))

print("\nHình 1:")
hinh1(n)

print("\nHình 2:")
hinh2(n)

print("\nHình 3:")
hinh3(n)
```

```python
#Câu 19
x = float(input("x = "))
n = int(input("n = "))

t = x
s = t

for k in range(0, n):
    denom = (2*(k+1))*(2*(k+1)+1)   # (2k+2)(2k+3)
    t = t * x * x / denom
    s += t

print("S(x,n) =", s)
```
