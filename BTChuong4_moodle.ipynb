```python
#Câu 1
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
#Câu 3
def BMI(height,weight):
 return weight/(height**2)
def PhanLoai(bmi):
 if bmi<18.5:
  return "Gầy"
 elif bmi<=24.9:
  return "Bình thường"
 elif bmi<=29.9:
  return "Hơi Béo"
 elif bmi<=34.9:
  return "Béo Phì Cấp Độ 1"
 elif bmi<=39.9:
  return "Béo Phì Cấp Độ 2"
 else:
  return "Béo Phì Cấp độ 3"
def NguyCoBenh(bmi):
 if bmi<18.5:
  return "Thấp"
 elif bmi<=24.9:
  return "Trung Bình"
 elif bmi<=29.9:
  return "Cao"
 elif bmi<=34.9:
  return "Cao"
 elif bmi<=39.9:
  return "Rất cao"
 else:
  return "Nguy Hiểm"
print("Nhập vào chiều cao:")
height=float(input())
print("Nhập vào cân nặng:")
weight=float(input())
bmi=BMI(height,weight)
print("BMI của bạn=",bmi)
print("Phân loại bạn=",PhanLoai(bmi))
print("Nguy cơ bệnh của Thím=",NguyCoBenh(bmi))
```

```python
#Câu 4
def ROI(dt,cp):
 return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
  return "Nên đầu tư"
 else:
  return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

```python
#Câu 7
import math

# Nhập tọa độ 2 điểm
xA, yA = map(float, input("Nhập tọa độ A (xA yA): ").split())
xB, yB = map(float, input("Nhập tọa độ B (xB yB): ").split())

# Tính độ dài đoạn AB
AB = math.sqrt((xB - xA)**2 + (yB - yA)**2)

print(f"Độ dài đoạn AB = {AB}")
```

```python
#Câu 13
def sum_divisors(n):
    """Tính tổng các ước số thực sự của n (không kể n)"""
    total = 1  # luôn có 1 là ước số
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            total += i
    return total if n > 1 else 0


def is_perfect(n):
    """Kiểm tra số hoàn thiện"""
    return sum_divisors(n) == n


def is_abundant(n):
    """Kiểm tra số thịnh vượng"""
    return sum_divisors(n) > n


# =========================
# Chạy thử
n = int(input("Nhập số nguyên dương n: "))

if is_perfect(n):
    print(f"{n} là số hoàn thiện (Perfect number).")
elif is_abundant(n):
    print(f"{n} là số thịnh vượng (Abundant number).")
else:
    print(f"{n} không phải số hoàn thiện, cũng không phải số thịnh vượng.")
```
