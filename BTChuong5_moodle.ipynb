```python

#Câu 4
# Một số hàm quan trọng trong xử lý chuỗi Python

text = "  Hello Python 123!  "

print("Chuỗi ban đầu:", text)

# Loại bỏ khoảng trắng ở đầu và cuối
print("strip():", text.strip())

# Chuyển hết sang chữ hoa
print("upper():", text.upper())

# Chuyển hết sang chữ thường
print("lower():", text.lower())

# Viết hoa chữ cái đầu
print("capitalize():", text.capitalize())

# Đếm số lần xuất hiện của ký tự/chuỗi
print("count('o'):", text.count('o'))

# Tìm vị trí xuất hiện của ký tự/chuỗi
print("find('Python'):", text.find("Python"))

# Kiểm tra chuỗi chỉ chứa chữ số?
print("isdigit():", "12345".isdigit())

# Thay thế chuỗi con
print("replace('Python', 'World'):", text.replace("Python", "World"))

# Tách chuỗi theo khoảng trắng
print("split():", text.split())

# Ghép chuỗi bằng ký tự '-'
print("join():", "-".join(["a", "b", "c"]))
```

```python
#Câu 5
def xu_ly_chuoi(s):
    # Khởi tạo các biến đếm
    so_in_hoa = 0
    so_in_thuong = 0
    so_chu_so = 0
    so_ky_tu_dac_biet = 0
    so_khoang_trang = 0
    so_nguyen_am = 0
    so_phu_am = 0

    nguyen_am = "aeiouAEIOU"

    for c in s:
        if c.isupper():
            so_in_hoa += 1
        elif c.islower():
            so_in_thuong += 1

        if c.isdigit():
            so_chu_so += 1
        elif not c.isalnum() and not c.isspace():
            so_ky_tu_dac_biet += 1
        elif c.isspace():
            so_khoang_trang += 1

        if c in nguyen_am:
            so_nguyen_am += 1
        elif c.isalpha() and c not in nguyen_am:
            so_phu_am += 1

    # Xuất kết quả
    print("Số chữ IN HOA:", so_in_hoa)
    print("Số chữ in thường:", so_in_thuong)
    print("Số chữ là chữ số:", so_chu_so)
    print("Số ký tự đặc biệt:", so_ky_tu_dac_biet)
    print("Số khoảng trắng:", so_khoang_trang)
    print("Số chữ nguyên âm:", so_nguyen_am)
    print("Số chữ phụ âm:", so_phu_am)


# Ví dụ chạy
chuoi = input("Nhập vào một chuỗi: ")
xu_ly_chuoi(chuoi)
```

```python

#Câu 6
import re

def NegativeNumberInStrings(s):
    # Tìm tất cả số nguyên âm (có dấu - phía trước)
    so_am = re.findall(r'-\d+', s)
    return [int(x) for x in so_am]


# Ví dụ chạy
chuoi = "abc-5xyz-12k9l--p"
print("Chuỗi:", chuoi)
print("Các số nguyên âm:", NegativeNumberInStrings(chuoi))
```

```python
#Câu 7
def toi_uu_chuoi_danh_tu(s):
    # Tách từ, loại bỏ khoảng trắng thừa
    tu = s.strip().split()
    # Viết hoa chữ cái đầu, còn lại viết thường
    tu_chuan = [word.capitalize() for word in tu]
    # Ghép lại thành 1 chuỗi
    return " ".join(tu_chuan)


# Ví dụ chạy
chuoi = "   TRần    duY    thAnH   "
print("Input :", chuoi)
print("Output:", toi_uu_chuoi_danh_tu(chuoi))
```

```python
#Câu 8
import os

# Hàm lấy tên file (bao gồm đuôi .mp3)
def lay_ten_file(path):
    return os.path.basename(path)

# Hàm lấy tên bài hát (không có đuôi)
def lay_ten_bai_hat(path):
    return os.path.splitext(os.path.basename(path))[0]


# Ví dụ chạy
duong_dan = "d:\\music\\muabui.mp3"
print("Đường dẫn:", duong_dan)
print("Tên file :", lay_ten_file(duong_dan))
print("Tên bài hát:", lay_ten_bai_hat(duong_dan))
```
