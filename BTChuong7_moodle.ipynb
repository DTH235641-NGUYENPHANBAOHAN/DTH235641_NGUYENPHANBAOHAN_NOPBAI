```python
#Câu 1
# Quản lý sản phẩm - lưu vào file text
# Tác giả: ChatGPT x An Ngo

FILENAME = "sanpham.txt"

# Hàm thêm sản phẩm vào file
def them_san_pham():
    print("\n=== NHẬP SẢN PHẨM ===")
    ma = input("Mã sản phẩm: ").strip()
    ten = input("Tên sản phẩm: ").strip()
    try:
        gia = float(input("Đơn giá: ").strip())
    except ValueError:
        print("Giá không hợp lệ, mặc định = 0")
        gia = 0.0

    # Tạo dòng dữ liệu
    dong = f"{ma};{ten};{gia}\n"

    # Ghi nối đuôi vào file
    with open(FILENAME, "a", encoding="utf-8") as f:
        f.write(dong)
    print("Đã lưu sản phẩm vào file!\n")


# Hàm đọc danh sách từ file
def doc_danh_sach():
    ds = []
    try:
        with open(FILENAME, "r", encoding="utf-8") as f:
            for line in f:
                parts = line.strip().split(";")
                if len(parts) == 3:
                    ma, ten, gia = parts
                    try:
                        gia = float(gia)
                    except ValueError:
                        gia = 0.0
                    ds.append({"ma": ma, "ten": ten, "gia": gia})
    except FileNotFoundError:
        print("Chưa có file dữ liệu!")
    return ds


# Hàm in danh sách
def xuat_danh_sach(ds):
    if not ds:
        print("Không có sản phẩm nào.")
        return
    print("\n=== DANH SÁCH SẢN PHẨM ===")
    print("{:<10} {:<20} {:>10}".format("Mã", "Tên", "Đơn giá"))
    print("-" * 42)
    for sp in ds:
        print("{:<10} {:<20} {:>10.2f}".format(sp["ma"], sp["ten"], sp["gia"]))


# Hàm sắp xếp giảm dần theo giá
def sap_xep_giam_dan(ds):
    return sorted(ds, key=lambda sp: sp["gia"], reverse=True)


# Menu chính
def menu():
    while True:
        print("\n===== QUẢN LÝ SẢN PHẨM =====")
        print("1. Nhập thêm sản phẩm")
        print("2. Xuất danh sách sản phẩm")
        print("3. Sắp xếp theo đơn giá giảm dần")
        print("0. Thoát")
        chon = input("Chọn chức năng: ").strip()

        if chon == "1":
            them_san_pham()
        elif chon == "2":
            ds = doc_danh_sach()
            xuat_danh_sach(ds)
        elif chon == "3":
            ds = doc_danh_sach()
            ds_sap = sap_xep_giam_dan(ds)
            xuat_danh_sach(ds_sap)
        elif chon == "0":
            print("Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại.")


# Chạy chương trình
if __name__ == "__main__":
    menu()
```

```python
#Câu 2
# Xử lý số trong Text File

# B1: Tạo file dữ liệu sẵn nếu chưa có
filename = "numbers.txt"

data = """5,6,8,9,-5
-9,5,4,7,8
6,7,8,3,6,46,7,2,-6,-7"""

# Tạo file (hoặc ghi đè lại cho chắc)
with open(filename, "w", encoding="utf-8") as f:
    f.write(data)


# B2: Đọc file và chuyển mỗi dòng thành list số
lists = []  # danh sách chứa các list dòng
with open(filename, "r", encoding="utf-8") as f:
    for line in f:
        # loại bỏ xuống dòng, tách theo dấu phẩy
        nums = [int(x) for x in line.strip().split(",")]
        lists.append(nums)

# Xuất từng dòng list ra màn hình
print("Danh sách các dòng (list):")
for i, lst in enumerate(lists, start=1):
    print(f"Dòng {i}: {lst}")

# B3: Xuất các số âm trên mỗi dòng
print("\nCác số âm trên từng dòng:")
for i, lst in enumerate(lists, start=1):
    am = [x for x in lst if x < 0]
    print(f"Dòng {i}: {am if am else 'Không có số âm'}")
```

```python
#Câu 7
import xlsxwriter
import os

# --- Tạo logo mẫu nếu chưa có ---
def tao_logo_mau():
    from PIL import Image, ImageDraw

    logo_path = "logo.png"
    if not os.path.exists(logo_path):
        img = Image.new("RGB", (200, 80), color=(0, 128, 255))
        d = ImageDraw.Draw(img)
        d.text((30, 25), "LOGO", fill=(255, 255, 255))
        img.save(logo_path)
        print("Đã tạo logo mẫu: logo.png")
    return logo_path


# --- Tạo file Excel ---
def tao_file_excel():
    file_name = "demo.xlsx"

    # Nếu đã có file, xóa để ghi mới
    if os.path.exists(file_name):
        os.remove(file_name)

    workbook = xlsxwriter.Workbook(file_name)
    worksheet = workbook.add_worksheet("SanPham")

    # Định dạng cột và hàng
    worksheet.set_column("A:E", 18)

    # Định dạng style
    bold_center = workbook.add_format({
        "bold": True,
        "align": "center",
        "valign": "vcenter",
        "border": 1,
        "bg_color": "#D9E1F2"
    })

    cell_center = workbook.add_format({
        "align": "center",
        "valign": "vcenter",
        "border": 1
    })

    cell_right = workbook.add_format({
        "align": "right",
        "valign": "vcenter",
        "border": 1
    })

    # --- Thêm logo ---
    logo_path = tao_logo_mau()
    worksheet.insert_image("A1", logo_path, {"x_scale": 0.5, "y_scale": 0.5})

    # --- Ghi tiêu đề bảng ---
    headers = ["STT", "MÃ SẢN PHẨM", "TÊN SẢN PHẨM", "SỐ LƯỢNG", "ĐƠN GIÁ"]
    worksheet.write_row("A5", headers, bold_center)

    # --- Dữ liệu sản phẩm ---
    data = [
        [1, "SP1", "Coca", 15, 15000],
        [2, "SP2", "Pepsi", 20, 18000],
        [3, "SP3", "Sting", 10, 12000],
    ]

    # --- Ghi dữ liệu ---
    row_start = 6
    for row_num, row_data in enumerate(data, start=row_start):
        worksheet.write(row_num - 1, 0, row_data[0], cell_center)
        worksheet.write(row_num - 1, 1, row_data[1], cell_center)
        worksheet.write(row_num - 1, 2, row_data[2], cell_center)
        worksheet.write(row_num - 1, 3, row_data[3], cell_center)
        worksheet.write(row_num - 1, 4, row_data[4], cell_right)

    # --- Tính tổng cộng ---
    worksheet.write(row_start + len(data) - 1, 3, "Tổng:", bold_center)
    worksheet.write_formula(
        row_start + len(data) - 1, 4,
        f"=SUM(E6:E{row_start + len(data) - 1})",
        bold_center
    )

    workbook.close()
    print(f"Đã tạo file Excel: {file_name}")


if __name__ == "__main__":
    tao_file_excel()
```

```python
#Câu 8
from openpyxl import load_workbook
import os
import xlsxwriter
from PIL import Image, ImageDraw

# --------------------------------------------------
# TẠO FILE EXCEL MẪU NẾU CHƯA CÓ (demo.xlsx)
# --------------------------------------------------
def tao_logo_mau():
    logo_path = "logo.png"
    if not os.path.exists(logo_path):
        img = Image.new("RGB", (200, 80), color=(0, 128, 255))
        d = ImageDraw.Draw(img)
        d.text((30, 25), "LOGO", fill=(255, 255, 255))
        img.save(logo_path)
        print("Đã tạo logo mẫu: logo.png")
    return logo_path


def tao_file_demo():
    if os.path.exists("demo.xlsx"):
        print("File demo.xlsx đã tồn tại, bỏ qua bước tạo.")
        return

    workbook = xlsxwriter.Workbook("demo.xlsx")
    worksheet = workbook.add_worksheet("SanPham")

    # Style
    bold_center = workbook.add_format({
        "bold": True, "align": "center", "valign": "vcenter",
        "border": 1, "bg_color": "#D9E1F2"
    })
    cell_center = workbook.add_format({
        "align": "center", "valign": "vcenter", "border": 1
    })
    cell_right = workbook.add_format({
        "align": "right", "valign": "vcenter", "border": 1
    })

    # Thêm logo
    logo_path = tao_logo_mau()
    worksheet.insert_image("A1", logo_path, {"x_scale": 0.5, "y_scale": 0.5})

    # Tiêu đề
    headers = ["STT", "MÃ SẢN PHẨM", "TÊN SẢN PHẨM", "SỐ LƯỢNG", "ĐƠN GIÁ"]
    worksheet.write_row("A5", headers, bold_center)

    # Dữ liệu mẫu
    data = [
        [1, "SP1", "Coca", 15, 15000],
        [2, "SP2", "Pepsi", 20, 18000],
        [3, "SP3", "Sting", 10, 12000],
    ]

    row_start = 6
    for row_num, row_data in enumerate(data, start=row_start):
        worksheet.write(row_num - 1, 0, row_data[0], cell_center)
        worksheet.write(row_num - 1, 1, row_data[1], cell_center)
        worksheet.write(row_num - 1, 2, row_data[2], cell_center)
        worksheet.write(row_num - 1, 3, row_data[3], cell_center)
        worksheet.write(row_num - 1, 4, row_data[4], cell_right)

    workbook.close()
    print("Đã tạo file demo.xlsx mẫu thành công!")


# --------------------------------------------------
# ĐỌC FILE EXCEL BẰNG openpyxl
# --------------------------------------------------
def doc_file_excel():
    # Tạo file nếu chưa có
    tao_file_demo()

    # Đọc file Excel
    wb = load_workbook("demo.xlsx")
    print("Các sheet có trong file:", wb.sheetnames)

    ws = wb[wb.sheetnames[0]]
    print("\nDỮ LIỆU TRONG SHEET:")

    for row in ws.values:
        for value in row:
            print(value, "\t", end="")
        print()

    wb.close()
    print("\nĐọc file Excel hoàn tất.")


# --------------------------------------------------
if __name__ == "__main__":
    doc_file_excel()
```

```python
#Câu 9
import os

FILE_NAME = "sanpham.txt"

# --- Tạo file nếu chưa có ---
def tao_file_text():
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, "w", encoding="utf-8") as f:
            f.write("MãSP|TênSP|ĐơnGiá|MãDanhMục|TênDanhMục\n")
        print(f"Đã tạo file {FILE_NAME} mới.")
    else:
        print("File đã tồn tại.")

# --- Lưu sản phẩm ---
def luu_sanpham(ma, ten, dongia, ma_dm, ten_dm):
    with open(FILE_NAME, "a", encoding="utf-8") as f:
        f.write(f"{ma}|{ten}|{dongia}|{ma_dm}|{ten_dm}\n")
    print(f"Đã lưu sản phẩm {ten}")

# --- Đọc danh sách sản phẩm ---
def doc_sanpham():
    if not os.path.exists(FILE_NAME):
        print("Chưa có file dữ liệu!")
        return []
    with open(FILE_NAME, "r", encoding="utf-8") as f:
        lines = f.readlines()[1:]  # bỏ dòng tiêu đề
        ds = []
        for line in lines:
            line = line.strip() # Remove leading/trailing whitespace
            if not line: # Skip empty lines
                continue
            parts = line.split("|")
            if len(parts) == 5: # Check if the line has 5 parts
                ma, ten, dongia, ma_dm, ten_dm = parts
                try:
                    ds.append({
                        "ma": ma,
                        "ten": ten,
                        "dongia": float(dongia),
                        "ma_dm": ma_dm,
                        "ten_dm": ten_dm
                    })
                except ValueError:
                    print(f"Bỏ qua dòng lỗi định dạng đơn giá: {line}")
            else:
                print(f"Bỏ qua dòng lỗi định dạng (không đủ 5 phần): {line}")

        return ds

# --- Hiển thị danh sách ---
def hien_thi(ds):
    if not ds:
        print("Không có sản phẩm nào.")
        return
    print("\nDANH SÁCH SẢN PHẨM:")
    print(f"{'Mã':<8}{'Tên':<15}{'Đơn giá':<10}{'Mã DM':<10}{'Tên DM'}")
    for sp in ds:
        print(f"{sp['ma']:<8}{sp['ten']:<15}{sp['dongia']:<10}{sp['ma_dm']:<10}{sp['ten_dm']}")
    print()

# --- Sắp xếp theo đơn giá ---
def sap_xep(ds):
    return sorted(ds, key=lambda x: x["dongia"])

# --- Tìm kiếm sản phẩm ---
def tim_kiem(ds, keyword):
    kq = [sp for sp in ds if keyword.lower() in sp["ten"].lower()]
    return kq

# --- Xóa sản phẩm ---
def xoa_sanpham(ds, ma):
    ds = [sp for sp in ds if sp["ma"] != ma]
    # Ghi lại file
    with open(FILE_NAME, "w", encoding="utf-8") as f:
        f.write("MãSP|TênSP|ĐơnGiá|MãDanhMục|TênDanhMục\n")
        for sp in ds:
            f.write(f"{sp['ma']}|{sp['ten']}|{sp['dongia']}|{sp['ma_dm']}|{sp['ten_dm']}\n")
    print("Đã xóa sản phẩm.")
    return ds

# --- MENU ---
def menu():
    tao_file_text()
    while True:
        print("""
===== QUẢN LÝ SẢN PHẨM =====
1. Thêm sản phẩm
2. Xem danh sách
3. Tìm sản phẩm theo tên
4. Sắp xếp theo đơn giá
5. Xóa sản phẩm
6. Thoát
============================
""")
        chon = input("Chọn: ")

        if chon == "1":
            ma = input("Mã SP: ")
            ten = input("Tên SP: ")
            dongia = float(input("Đơn giá: "))
            ma_dm = input("Mã danh mục: ")
            ten_dm = input("Tên danh mục: ")
            luu_sanpham(ma, ten, dongia, ma_dm, ten_dm)

        elif chon == "2":
            hien_thi(doc_sanpham())

        elif chon == "3":
            kw = input("Nhập tên cần tìm: ")
            kq = tim_kiem(doc_sanpham(), kw)
            hien_thi(kq)

        elif chon == "4":
            ds = sap_xep(doc_sanpham())
            hien_thi(ds)

        elif chon == "5":
            ma = input("Nhập mã SP cần xóa: ")
            ds = doc_sanpham()
            xoa_sanpham(ds, ma)

        elif chon == "6":
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ!")

if __name__ == "__main__":
    menu()
```

```python
#Câu 10
import json
import os

FILE_NAME = "sinhvien.json"

# --- Tạo file JSON nếu chưa có ---
def tao_file_json():
    if not os.path.exists(FILE_NAME):
        with open(FILE_NAME, "w", encoding="utf-8") as f:
            json.dump({"DSLop": []}, f, ensure_ascii=False, indent=4)
        print(f"Đã tạo file {FILE_NAME} mới.")
    else:
        print("File JSON đã tồn tại.")

# --- Đọc dữ liệu ---
def doc_du_lieu():
    with open(FILE_NAME, "r", encoding="utf-8") as f:
        return json.load(f)

# --- Ghi dữ liệu ---
def ghi_du_lieu(data):
    with open(FILE_NAME, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

# --- Thêm lớp học ---
def them_lop(ma, ten):
    data = doc_du_lieu()
    data["DSLop"].append({"MaLop": ma, "TenLop": ten, "SinhVien": []})
    ghi_du_lieu(data)
    print(f"Đã thêm lớp {ten}")

# --- Thêm sinh viên ---
def them_sv(ma_lop, ma_sv, ten, namsinh):
    data = doc_du_lieu()
    for lop in data["DSLop"]:
        if lop["MaLop"] == ma_lop:
            lop["SinhVien"].append({"Ma": ma_sv, "Ten": ten, "NamSinh": namsinh})
            ghi_du_lieu(data)
            print(f"Đã thêm sinh viên {ten} vào lớp {ma_lop}")
            return
    print("Không tìm thấy lớp!")

# --- Hiển thị toàn bộ ---
def hien_thi():
    data = doc_du_lieu()
    for lop in data["DSLop"]:
        print(f"\nLớp: {lop['TenLop']} ({lop['MaLop']})")
        for sv in lop["SinhVien"]:
            print(f"  - {sv['Ma']}: {sv['Ten']} ({sv['NamSinh']})")
    print()

# --- Tìm kiếm sinh viên ---
def tim_sv(keyword):
    data = doc_du_lieu()
    kq = []
    for lop in data["DSLop"]:
        for sv in lop["SinhVien"]:
            if keyword.lower() in sv["Ten"].lower():
                kq.append((lop["MaLop"], sv))
    if not kq:
        print("Không tìm thấy sinh viên.")
        return
    print("\nKết quả tìm kiếm:")
    for lop, sv in kq:
        print(f"Lớp {lop} - {sv['Ma']}: {sv['Ten']} ({sv['NamSinh']})")

# --- Xóa sinh viên ---
def xoa_sv(ma_sv):
    data = doc_du_lieu()
    for lop in data["DSLop"]:
        lop["SinhVien"] = [sv for sv in lop["SinhVien"] if sv["Ma"] != ma_sv]
    ghi_du_lieu(data)
    print("Đã xóa sinh viên.")

# --- Menu chính ---
def menu():
    tao_file_json()
    while True:
        print("""
===== QUẢN LÝ SINH VIÊN =====
1. Thêm lớp học
2. Thêm sinh viên
3. Xem danh sách
4. Tìm sinh viên theo tên
5. Xóa sinh viên
6. Thoát
=============================
""")
        chon = input("Chọn: ")

        if chon == "1":
            ma = input("Mã lớp: ")
            ten = input("Tên lớp: ")
            them_lop(ma, ten)

        elif chon == "2":
            ma_lop = input("Nhập mã lớp: ")
            ma_sv = input("Mã sinh viên: ")
            ten = input("Tên sinh viên: ")
            namsinh = int(input("Năm sinh: "))
            them_sv(ma_lop, ma_sv, ten, namsinh)

        elif chon == "3":
            hien_thi()

        elif chon == "4":
            kw = input("Nhập tên cần tìm: ")
            tim_sv(kw)

        elif chon == "5":
            ma_sv = input("Nhập mã SV cần xóa: ")
            xoa_sv(ma_sv)

        elif chon == "6":
            print("Thoát chương trình.")
            break

        else:
            print("Lựa chọn không hợp lệ!")

if __name__ == "__main__":
    menu()
```

```python
#Câu 11
import openpyxl
from openpyxl import Workbook
import os

# Tên file Excel
FILE_NAME = "nhanvien.xlsx"

# --- Hàm tạo file Excel nếu chưa có ---
def tao_file_excel():
    if not os.path.exists(FILE_NAME):
        wb = Workbook()
        ws = wb.active
        ws.title = "NhanVien"
        ws.append(["STT", "Mã", "Tên", "Tuổi"])
        wb.save(FILE_NAME)
        print(f"Đã tạo file Excel mới: {FILE_NAME}")
    else:
        print("File Excel đã tồn tại.")

# --- Hàm thêm nhân viên vào Excel ---
def them_nhan_vien(ma, ten, tuoi):
    wb = openpyxl.load_workbook(FILE_NAME)
    ws = wb.active
    stt = ws.max_row  # Tự động tăng STT
    ws.append([stt, ma, ten, tuoi])
    wb.save(FILE_NAME)
    print(f"Đã thêm nhân viên {ten} vào file Excel!")

# --- Hàm đọc danh sách nhân viên ---
def doc_danh_sach():
    wb = openpyxl.load_workbook(FILE_NAME)
    ws = wb.active
    print("\nDANH SÁCH NHÂN VIÊN:")
    for row in ws.iter_rows(values_only=True):
        print(row)
    print()

# --- Hàm sắp xếp nhân viên theo tuổi tăng dần ---
def sap_xep_theo_tuoi():
    wb = openpyxl.load_workbook(FILE_NAME)
    ws = wb.active

    # Lấy dữ liệu bỏ qua dòng tiêu đề
    data = list(ws.iter_rows(values_only=True))
    tieu_de = data[0]
    nhanvien = data[1:]

    # Sắp xếp theo tuổi
    nhanvien.sort(key=lambda x: x[3])

    # Ghi lại vào file
    ws.delete_rows(2, ws.max_row)
    for i, nv in enumerate(nhanvien, start=1):
        ws.append([i, nv[1], nv[2], nv[3]])

    wb.save(FILE_NAME)
    print("Đã sắp xếp nhân viên theo tuổi tăng dần!")

# --- Menu chính ---
def menu():
    tao_file_excel()

    while True:
        print("""
===== QUẢN LÝ NHÂN VIÊN =====
1. Thêm nhân viên
2. Xem danh sách nhân viên
3. Sắp xếp theo tuổi tăng dần
4. Thoát
=============================
""")
        chon = input("Chọn chức năng: ")

        if chon == "1":
            ma = input("Nhập mã nhân viên: ")
            ten = input("Nhập tên nhân viên: ")
            tuoi = int(input("Nhập tuổi: "))
            them_nhan_vien(ma, ten, tuoi)

        elif chon == "2":
            doc_danh_sach()

        elif chon == "3":
            sap_xep_theo_tuoi()

        elif chon == "4":
            print("Thoát chương trình.")
            break

        else:
            print("Lựa chọn không hợp lệ, vui lòng nhập lại!")

# --- Chạy chương trình ---
if __name__ == "__main__":
    menu()
```

```python
#Câu 12
import csv
import random
import os

def tao_file_csv(ten_file):
    """Tạo file CSV với 10 dòng, mỗi dòng có 10 số ngẫu nhiên (1–100)."""
    with open(ten_file, mode='w', newline='') as f:
        writer = csv.writer(f, delimiter=';')
        for _ in range(10):  # 10 dòng
            dong = [random.randint(1, 100) for _ in range(10)]  # 10 số ngẫu nhiên
            writer.writerow(dong)
    print(f"Đã tạo file dữ liệu: {ten_file}")

def doc_va_tinh_tong(ten_file):
    """Đọc file CSV và in tổng từng dòng."""
    with open(ten_file, mode='r') as f:
        reader = csv.reader(f, delimiter=';')
        print("\nKết quả tính tổng mỗi dòng:\n")
        for i, dong in enumerate(reader, start=1):
            # Bỏ qua các phần tử rỗng (nếu có do dấu ';' cuối dòng)
            so_hop_le = [int(x) for x in dong if x.strip() != '']
            tong = sum(so_hop_le)
            print(f"Dòng {i}: {so_hop_le} → Tổng = {tong}")

def main():
    ten_file = "dulieu_ngau_nhien.csv"

    # Nếu file chưa tồn tại thì tự động tạo
    if not os.path.exists(ten_file):
        tao_file_csv(ten_file)
    else:
        print(f"File '{ten_file}' đã tồn tại, sử dụng lại dữ liệu cũ.")

    # Đọc và xử lý file
    doc_va_tinh_tong(ten_file)

if __name__ == "__main__":
    main()
```

```python
#Câu 13
import xml.etree.ElementTree as ET
import os

# === TẠO FILE XML MẪU NẾU CHƯA CÓ ===
def tao_file_mau():
    # File nhóm thiết bị
    nhom_file = "nhomthietbi.xml"
    if not os.path.exists(nhom_file):
        root = ET.Element("nhoms")

        data_nhoms = [
            ("n1", "Nhóm 1 - Thiết bị điện"),
            ("n2", "Nhóm 2 - Thiết bị văn phòng"),
            ("n3", "Nhóm 3 - Thiết bị mạng")
        ]

        for ma, ten in data_nhoms:
            nhom = ET.SubElement(root, "nhom")
            ET.SubElement(nhom, "ma").text = ma
            ET.SubElement(nhom, "ten").text = ten

        tree = ET.ElementTree(root)
        tree.write(nhom_file, encoding="UTF-8", xml_declaration=True)
        print(f"Đã tạo file mẫu: {nhom_file}")

    # File thiết bị
    thietbi_file = "thietbi.xml"
    if not os.path.exists(thietbi_file):
        root = ET.Element("thietbis")

        data_thietbi = [
            ("n1", "tb1", "Thiết bị 1 - Bóng đèn"),
            ("n1", "tb2", "Thiết bị 2 - Quạt điện"),
            ("n2", "tb3", "Thiết bị 3 - Máy in"),
            ("n3", "tb4", "Thiết bị 4 - Router WiFi"),
            ("n3", "tb5", "Thiết bị 5 - Switch mạng")
        ]

        for manhom, ma, ten in data_thietbi:
            tb = ET.SubElement(root, "thietbi", {"manhom": manhom})
            ET.SubElement(tb, "ma").text = ma
            ET.SubElement(tb, "ten").text = ten

        tree = ET.ElementTree(root)
        tree.write(thietbi_file, encoding="UTF-8", xml_declaration=True)
        print(f"Đã tạo file mẫu: {thietbi_file}")

# === Đọc dữ liệu từ file XML ===
def load_nhomthietbi(filename):
    tree = ET.parse(filename)
    root = tree.getroot()
    ds_nhom = []
    for nhom in root.findall('nhom'):
        ma = nhom.find('ma').text
        ten = nhom.find('ten').text
        ds_nhom.append({'ma': ma, 'ten': ten})
    return ds_nhom

def load_thietbi(filename):
    tree = ET.parse(filename)
    root = tree.getroot()
    ds_tb = []
    for tb in root.findall('thietbi'):
        manhom = tb.get('manhom')
        ma = tb.find('ma').text
        ten = tb.find('ten').text
        ds_tb.append({'manhom': manhom, 'ma': ma, 'ten': ten})
    return ds_tb

# === Hiển thị danh sách nhóm ===
def hien_thi_nhom(ds_nhom):
    print("\nDanh sách nhóm thiết bị:")
    for n in ds_nhom:
        print(f"  - Mã nhóm: {n['ma']}, Tên nhóm: {n['ten']}")

# === Hiển thị toàn bộ thiết bị ===
def hien_thi_thietbi(ds_tb):
    print("\nDanh sách thiết bị:")
    for tb in ds_tb:
        print(f"  - Mã: {tb['ma']}, Tên: {tb['ten']}, Thuộc nhóm: {tb['manhom']}")

# === Lọc thiết bị theo nhóm ===
def loc_thietbi_theo_nhom(ds_tb, manhom):
    return [tb for tb in ds_tb if tb['manhom'] == manhom]

# === Tìm nhóm có nhiều thiết bị nhất ===
def nhom_nhieu_tb_nhat(ds_tb, ds_nhom):
    dem = {}
    for tb in ds_tb:
        dem[tb['manhom']] = dem.get(tb['manhom'], 0) + 1
    max_sl = max(dem.values())
    nhom_max = [ma for ma, sl in dem.items() if sl == max_sl]
    ten_nhom_max = [n['ten'] for n in ds_nhom if n['ma'] in nhom_max]
    return ten_nhom_max, max_sl

# === Thêm thiết bị mới và lưu file ===
def them_thietbi(filename, ds_tb):
    manhom = input("Nhập mã nhóm (vd: n1, n2...): ")
    ma = input("Nhập mã thiết bị mới: ")
    ten = input("Nhập tên thiết bị mới: ")

    root = ET.parse(filename).getroot()
    new_tb = ET.Element("thietbi", {"manhom": manhom})
    ET.SubElement(new_tb, "ma").text = ma
    ET.SubElement(new_tb, "ten").text = ten
    root.append(new_tb)

    tree = ET.ElementTree(root)
    tree.write(filename, encoding="UTF-8", xml_declaration=True)
    print("Đã thêm thiết bị mới và lưu lại file XML.")

# === Chương trình chính ===
def main():
    tao_file_mau()  # Tự động tạo file mẫu nếu chưa có

    nhom_file = "nhomthietbi.xml"
    thietbi_file = "thietbi.xml"

    ds_nhom = load_nhomthietbi(nhom_file)
    ds_tb = load_thietbi(thietbi_file)

    while True:
        print("\n===QUẢN LÝ THIẾT BỊ ===")
        print("1. Hiển thị danh sách nhóm thiết bị")
        print("2. Hiển thị toàn bộ thiết bị")
        print("3. Lọc thiết bị theo nhóm")
        print("4. Xuất nhóm có nhiều thiết bị nhất")
        print("5. Thêm thiết bị mới và lưu file XML")
        print("0. Thoát")
        chon = input("Chọn chức năng: ")

        if chon == "1":
            hien_thi_nhom(ds_nhom)
        elif chon == "2":
            hien_thi_thietbi(ds_tb)
        elif chon == "3":
            ma = input("Nhập mã nhóm cần lọc (vd: n1, n2...): ")
            kq = loc_thietbi_theo_nhom(ds_tb, ma)
            if kq:
                print(f"\nThiết bị thuộc nhóm {ma}:")
                for tb in kq:
                    print(f"  - {tb['ma']}: {tb['ten']}")
            else:
                print("Không có thiết bị nào thuộc nhóm này.")
        elif chon == "4":
            ten_nhom_max, sl = nhom_nhieu_tb_nhat(ds_tb, ds_nhom)
            print("\nNhóm có nhiều thiết bị nhất:")
            for ten in ten_nhom_max:
                print(f"  - {ten} ({sl} thiết bị)")
        elif chon == "5":
            them_thietbi(thietbi_file, ds_tb)
            ds_tb = load_thietbi(thietbi_file)  # Cập nhật lại danh sách sau khi thêm
        elif chon == "0":
            print("Kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ!")

if __name__ == "__main__":
    main()
```
