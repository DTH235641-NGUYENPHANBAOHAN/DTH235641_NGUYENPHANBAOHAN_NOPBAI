```python
width = 6   # chiều ngang
height = 4  # chiều dọc

for i in range(height):
    for j in range(width):
        # nếu là dòng đầu, dòng cuối, hoặc cột đầu, cột cuối -> in '*'
        if i == 0 or i == height - 1 or j == 0 or j == width - 1:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
width = 9
height = 4

for _ in range(height):
    print('*' * width)
```

```python
h = [1, 3, 5, 3, 7, 9]
max_width = max(h)
for sao in h:
    space = (max_width - sao) // 2
    print(" " * space + "*" * sao)
for _ in range(2):
    space = (max_width - 3) // 2
    print(" " * space + "*" * 3)
```
